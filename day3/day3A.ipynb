```python
import pandas
import numpy
```

```python
with open('day3input.txt') as inputfile:
    data = inputfile.read().splitlines()
```

```python
data
```

```python
def get_data_dict(data):
    d = {}
    for claim in data:
        id = claim.split()[0]
        left, top = claim.split()[2].strip(':').split(',')
        width, height = claim.split()[3].split('x')
        d[id] = (int(left), int(top), int(width), int(height))
    return d
    
```

```python
def update_df(df, claim):
    left, top, width, height = claim
    for col in range(left, left + width):
        for row in range(top, top + height):
            df[col][row] += 1
    return df
    
```

```python
def get_occupied(df):
    total = df.shape[0] * df.shape[1]
    listvals = [list(val) for val in df.values]
    flatvals = sum(listvals, [])
    occupied_by_more_than_one = [v for v in flatvals if v > 1]
    return len(occupied_by_more_than_one)
```

```python
def get_unique(data_dict, df):
    for id, claim in data_dict.items():
        left, top, width, height = claim
        left, top, width, height = claim
        unique = True
        for col in range(left, left + width):
            for row in range(top, top + height):
                if df[col][row] != 1:
                    unique = False
        if unique == True:
            return id
```

```python
test = ['#1 @ 1,3: 4x4', '#2 @ 3,1: 4x4', '#3 @ 5,5: 2x2']
```

```python
test_data_dict = get_data_dict(test)
```

```python
test_df = pandas.DataFrame(numpy.zeros((9, 11)))
for claim in test_data_dict.values():
    update_df(test_df, claim)
```
