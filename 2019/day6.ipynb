```python
Orbit = namedtuple("Orbit", ["name", "direct", "num_indirect"])
```

```python
def process_initial_input(test_input):
    indirect_dict = {}
    orbits = []
    for item in test_input:
        direct, name = item.split(")")
        if direct == "COM":
            num_indirect = 0
            indirect_dict.setdefault(0, []).append(Orbit(name=name, direct=direct, num_indirect=num_indirect))
        else:
            num_indirect = None
            orbits.append(Orbit(name=name, direct=direct, num_indirect=num_indirect))
    return indirect_dict, orbits
```

```python
def find_indirect_orbits(orbit_list, indirect_dict, counter=1):
    processed_orbit_list = 1 # indirect_dict[0]
    while processed_orbit_list <= len(orbit_list):
        for orbit in orbit_list:
            if orbit.direct in [orb.name for orb in indirect_dict[counter-1]]:
                indirect_dict.setdefault(counter, []).append(orbit)
                processed_orbit_list += 1
        counter += 1
    return indirect_dict
    
```

```python
def counts(indirect_dict):
    direct_count = 0
    indirect_count = 0
    for num_indirects, orbits in indirect_dict.items():
        indirect_count += (num_indirects * len(orbits))
        direct_count += len(orbits)
    return indirect_count + direct_count
```

```python
indirect_count, orbits = process_initial_input(read_input)
```

```python
indirect_dict = find_indirect_orbits(orbits, indirect_count)
```

```python
def find_path(start):
    orb_path = []
    current = start
    while current:
        next_orb = next((orb for orb in orbits if orb.name == current.direct), None)
        if next_orb is not None:
            orb_path.append(next_orb)
        current = next_orb
    return orb_path
```
