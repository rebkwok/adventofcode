```python
import re
```

```python
input_file = "input/day7.txt"
with open(input_file) as input:
    read_data = input.read().strip().split("\n")
    read_data = [re.sub(r'bags\.|bag\.|bags|bag', '', item).strip() for item in read_data]
    bags_and_content = [rule.split(" contain ") for rule in read_data]
    bags_and_content = {key.strip(): value.strip() for key, value in bags_and_content}
    bags_dict = {}
    for key, value in bags_and_content.items():
        if value.strip() == "no other":
            bags_dict[key] = {"none": 0}
        else:
            items = value.split(',')
            for item in items:
                number = re.match(r"^[0-9]+", item.strip()).group()
                description = item.replace(number, '').strip()
                bags_dict.setdefault(key, {})[description] = int(number)
                
```

```python
def superbags(target_bags, found_superbags=None):
    if not target_bags:
        return found_superbags
    found_superbags = found_superbags or set()
    new_targets = set()
    for target_bag in target_bags:
        found = {key for key, value in bags_dict.items() if target_bag in value}
        new_targets = new_targets.union(found)
        found_superbags = found_superbags.union(found)
    return superbags(new_targets, found_superbags)
    
```

```python
containing_bags = superbags(["shiny gold"])
```

```python
def inner_bags(target_bags, contents=None):
    print(target_bags)
    if not target_bags:
        return contents
    contents = contents or 0
    new_targets = []
    for target_bag in target_bags:
        for bag, number in bags_dict[target_bag].items():
            if bag != "none":
                contents += number
                new_targets += [bag] * number
    # print(new_targets, contents)
    return inner_bags(new_targets, contents)
    
    
```
