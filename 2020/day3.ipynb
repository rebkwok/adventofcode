```python
from functools import reduce
import operator
```

```python
input_file = "input/day3.txt"
with open(input_file) as input:
    read_data = input.read().strip().split("\n")
```

```python
read_data
```

```python
def find_trees(right_shift, down_shift):
    row = col = trees = 0
    while row < len(read_data):
        current_row = read_data[row]
        if col >= len(read_data[row]):
            while len(current_row) <= col:
                current_row += read_data[row]
        check_position = current_row[col]
        if check_position == '#':
            trees += 1
        col += right_shift
        row += down_shift
    return trees
```

```python

trees = [find_trees(right_shift, down_shift) for (right_shift, down_shift) in [(1, 1), (3, 1), (5, 1), (7, 1), (1, 2)]]
    
```

```python
trees
```

```python
reduce(operator.mul, trees, 1)
```
