```python
from collections import Counter
```

```python
input_file = "input/day6.txt"
with open(input_file) as input:
    read_data = input.read().strip().split("\n\n")
    group_answers = [data.split("\n") for data in read_data]
    combined_answers = [''.join(data) for data in group_answers]
```

```python
read_data
```

```python
# Part 1
sum([len(set(data)) for data in combined_answers])
```

```python
# part 2
split_group_answers = [
    [[answer for answer in answers] for answers in individual_answers] for individual_answers in group_answers
]
    
```

```python
split_group_answers
```

```python
group_yes = []
for group in split_group_answers:
    unique = set(sum(group, []))
    all_yes = []
    for item in unique:
        counts = sum(1 for individual in group if item in individual)
        if counts == len(group):
            all_yes.append(item)
    group_yes.append(all_yes)
                      
    
```

```python
group_yes
```

```python
sum(len(yes) for yes in group_yes)
```
