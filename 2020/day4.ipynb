```python
import re
```

```python
input_file = "input/day4.txt"
collated_passports = []
with open(input_file) as input:
    read_data = input.read().strip().split("\n")
    segment = []
    for i, row in enumerate(read_data):
        if row != '':
            segment.append(row)
            if i == (len(read_data) - 1):
                collated_passports.append(tuple(segment))
        else:
            collated_passports.append(tuple(segment))
            segment = []

passports = [tuple(' '.join(passport).split()) for passport in collated_passports]
passports = [dict([item.split(':') for item in p]) for p in passports]
```

```python
required_fields = {"byr", "iyr", "eyr", "hgt", "hcl", "ecl", "pid"}
```

```python
def test_year(value, min_year, max_year):
    if len(value) != 4:
        return False
    try:
        value = int(value)
    except ValueError:
        return False
    if not (min_year <= value <= max_year):
        return False
    return True
```

```python
def test_height(value):
    if len(value) < 3:
        return False
    try:
        height = int(value[:-2])
    except ValueError:
        return False
    units = value[-2:]
    if units not in ["cm", "in"]:
        return False
    if units == "cm":
        if not (150 <= height <= 193):
            return False
    elif units == "in":
        if not (59 <= height <= 76):
            return False
    return True
    
```

```python
def is_valid(passport):
    missing = required_fields - set(passport.keys())
    if missing:
        return False
    
    if not test_year(passport["byr"], 1920, 2002):
        return False
    
    if not test_year(passport["iyr"], 2010, 2020):
        return False
    
    if not test_year(passport["eyr"], 2020, 2030):
        return False
    
    if not test_height(passport["hgt"]):
        return False
    
    if not re.match(r"#[a-f0-9]{6}$", passport["hcl"]):
        return False
    
    if not (len(passport["ecl"]) == 3 and re.match(r"amb|blu|brn|gry|grn|hzl|oth", passport["ecl"])):
        return False
    
    if not re.match(r"^[0-9]{9}$", passport["pid"]):
        return False
    
    return True
```

```python
valid = 0
for passport in passports:
    if is_valid(passport):
        valid += 1
```
