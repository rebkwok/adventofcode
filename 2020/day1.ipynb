```python
from itertools import combinations
from functools import reduce
import operator
```

```python
input_file = "input/day1.txt"
with open(input_file) as input:
    data = [int(input_data) for input_data in input.read().strip().split("\n")]
    
```

```python
data
```

```python
def find_combination(number_of_items):
    data_combos = combinations(data, number_of_items)
    for combo in data_combos:
        if sum(combo) == 2020:
            print(combo)
            print(reduce(operator.mul, combo, 1))
            break
```

```python
find_combination(3)
```
