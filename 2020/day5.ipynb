```python
def decode(input_data, code, index, end_index, lower_code):
    if index == end_index:
        if code[index] == lower_code:
            return input_data[0]
        return input_data[1]
    half_rows = int(len(input_data) / 2)
    if code[index] == lower_code:
        input_data = input_data[:half_rows]
    else:
        input_data = input_data[half_rows:]
    return decode(input_data, code, index + 1, end_index, lower_code)
```

```python
def find_seat(code):
    rows = list(range(128))
    columns = list(range(8))
    row = decode(rows, code[:7], 0, 6, "F")
    column = decode(columns, code[7:], 0, 2, "L")
    seat_id = row * 8 + column
    return row, column, seat_id
```
