```python
from itertools import combinations
```

```python
input_file = "input/day9.txt"
with open(input_file) as input:
    read_data = input.read().strip().split("\n")
    read_data = [int(x) for x in read_data]
```

```python
read_data
```

```python
# part 1
def check_data(input_data, preamble_length):
    for i in range(preamble_length, len(input_data)):
        test_value = input_data[i]
        preamble = input_data[i - preamble_length:i]
        data = read_data[i:]
        preamble_combinations = list(combinations(preamble, 2))
        valid_numbers = [sum(combo) for combo in preamble_combinations]
        if test_value not in valid_numbers:
            return test_value
    
```

```python
preamble_length = 25
target_value = check_data(read_data, preamble_length)
```

```python
# part 2
def find_sequence(read_data, target_value):
    for i in range(len(read_data)):
        for j in range(i + 1, len(read_data)):
            sequence = read_data[i:j+1]
            sequence_total = sum(sequence)
            if sequence_total == target_value:
                return sequence
            if sequence_total > target_value:
                break
        
```

```python
target_value
```

```python
found = find_sequence(read_data, target_value)
```
